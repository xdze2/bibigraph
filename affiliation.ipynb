```python
affiliationsFlatten = {  aff.strip() for article in biblio.articles for aff in article.affiliation  }
print(len(affiliationsFlatten))
```

```python
def getCountry( aff ):
    country = aff.split(',')[-1].strip()
    return country
```

```python
affiliationsByCountry = {}

for aff in affiliationsFlatten:
    country = getCountry( aff )

    if country in affiliationsByCountry:
        affiliationsByCountry[country].append(aff)
    else:
        affiliationsByCountry[country] = [ aff ]
```

## Try to match the authors with an affiliation

```python
auth2aff = {}

for article in biblio.articles:
        
    for auth in article.authors:
        
        if auth not in auth2aff:
            auth2aff[auth] = []
            
        for aff in article.affiliation:
            aff  = aff.strip()
            auth2aff[auth].append( aff )

        
```

```python
for auth, affs in auth2aff.items():
    
    auth2aff[auth] = countList( affs )
```

```python
list( auth2aff.items() )[10]
```

```python
len( auth2aff )
```

```python
def countList( A ):
    dico = {}
    
    for a in A:
        if a in dico:
            dico[a] += 1
        else:
            dico[a] = 1
            
    return sorted( dico.items(), key= lambda x:x[1] )
    
```
